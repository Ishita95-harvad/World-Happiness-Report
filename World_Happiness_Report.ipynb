## Data loading

### Subtask:
Load the dataset "2015.csv" into a pandas DataFrame.


**Reasoning**:
Load the dataset "2015.csv" into a pandas DataFrame and display its first few rows and shape.



```python
import pandas as pd

try:
    df = pd.read_csv('2015.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: '2015.csv' not found.")
    df = None
except pd.errors.ParserError:
    print("Error: Unable to parse '2015.csv'. Check file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded data to understand its structure, identify missing values, and examine the distribution of key variables.


**Reasoning**:
I need to explore the data by examining the first few rows, data types, missing values, descriptive statistics, and distributions of key variables.  I'll also check for potential relationships between variables.



```python
# Examine the first few rows
display(df.head())

# Determine data types
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Summarize descriptive statistics
display(df.describe())

# Explore distributions of key variables
import matplotlib.pyplot as plt

# Histograms for numerical features
plt.figure(figsize=(12, 6))
plt.hist(df['Happiness Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(df['Economy (GDP per Capita)'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution of Economy (GDP per Capita)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Frequency')
plt.show()

# Bar plot for a categorical feature
region_counts = df['Region'].value_counts()
plt.figure(figsize=(12, 6))
region_counts.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Regions')
plt.xlabel('Region')
plt.ylabel('Frequency')
plt.show()

# Scatter plot for numerical pairs
plt.figure(figsize=(10, 6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'], color='purple')
plt.title('Relationship between Economy and Happiness Score')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.show()
```

## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and outliers.


**Reasoning**:
Inspect data types, convert incorrect data types, handle outliers, and verify the cleaning process.



```python
# Inspect data types
print(df.dtypes)

# Convert data types if necessary (no incorrect types found in initial exploration)
# ... (No conversion needed based on previous exploration)

# Handle outliers (using IQR method for numerical features)
numerical_cols = ['Happiness Score', 'Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# Verify data cleaning
print(df.describe())

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    plt.hist(df[col], bins=20)
    plt.title(col)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    plt.boxplot(df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

## Data analysis

### Subtask:
Perform descriptive statistics and identify trends or patterns in the data.


**Reasoning**:
Calculate summary statistics, correlation matrix, analyze regional happiness, and perform a t-test to compare happiness scores of two regions.



```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Summary statistics
numerical_cols = ['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual', 'Happiness_Economy_Diff']
display(df[numerical_cols].describe())

# 2. Correlation matrix
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)
print("Correlation with Happiness Score:\n", correlation_matrix['Happiness Score'].sort_values(ascending=False))


# 3. Analyze regional happiness
region_happiness = df.groupby('Region')['Happiness Score'].mean()
plt.figure(figsize=(12, 6))
region_happiness.plot(kind='bar', color='skyblue')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 4. T-test (example: comparing Western Europe and Sub-Saharan Africa)
western_europe_happiness = df[df['Region'] == 'Western Europe']['Happiness Score']
sub_saharan_africa_happiness = df[df['Region'] == 'Sub-Saharan Africa']['Happiness Score']

t_statistic, p_value = stats.ttest_ind(western_europe_happiness, sub_saharan_africa_happiness)
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a statistically significant difference in happiness scores between Western Europe and Sub-Saharan Africa.")
else:
    print("There is no statistically significant difference in happiness scores between Western Europe and Sub-Saharan Africa.")
```

Visualize the key findings from the data analysis.

Create the visualizations specified in the instructions: a histogram for 'Happiness Score', a scatter plot of 'Happiness Score' vs. 'Economy (GDP per Capita)', a box plot of 'Happiness Score' by 'Region', and a heatmap of the correlation matrix.
